```python
DOSYA_ADI = "gorevler.txt"

ornek_gorevler = [
    "Python ödevini bitir",
    "Matematik tekrar yap",
    "CV dosyasını düzenle",
    "Sunum slaytlarını hazırla",
    "Arkadaşla proje toplantısı yap"
]

def gorevleri_kaydet(gorevler):
    with open(DOSYA_ADI, "w", encoding="utf-8") as dosya:
        for gorev in gorevler:
            dosya.write(gorev + "\n")

def gorevleri_oku():
    try:
        with open(DOSYA_ADI, "r", encoding="utf-8") as dosya:
            return [satir.strip() for satir in dosya if satir.strip()]
    except FileNotFoundError:
        gorevleri_kaydet(ornek_gorevler)
        return ornek_gorevler

def gorevleri_listele(gorevler):
    if not gorevler:
        print("Hiç görev yok.")
    else:
        print("\nGörev Listesi:")
        for i, gorev in enumerate(gorevler, start=1):
            print(f"{i}. {gorev}")

def gorev_ekle(gorevler):
    yeni = input("Yeni görev: ").strip()
    if yeni == "":
        print("Boş görev eklenemez.")
    else:
        gorevler.append(yeni)
        gorevleri_kaydet(gorevler)
        print("Görev eklendi.")

def gorev_sil(gorevler):
    gorevleri_listele(gorevler)
    if not gorevler:
        return
    try:
        numara = int(input("Silinecek görev numarası: "))
        if 1 <= numara <= len(gorevler):
            silinen = gorevler.pop(numara - 1)
            gorevleri_kaydet(gorevler)
            print(f"'{silinen}' silindi.")
        else:
            print("Geçersiz numara.")
    except ValueError:
        print("Lütfen sayı girin.")

def gorev_duzenle(gorevler):
    gorevleri_listele(gorevler)
    if not gorevler:
        return
    try:
        numara = int(input("Düzenlenecek görev numarası: "))
        if 1 <= numara <= len(gorevler):
            yeni = input("Yeni görev metni: ").strip()
            if yeni == "":
                print("Boş görev olamaz.")
                return
            gorevler[numara - 1] = yeni
            gorevleri_kaydet(gorevler)
            print("Görev güncellendi.")
        else:
            print("Geçersiz numara.")
    except ValueError:
        print("Lütfen sayı girin.")

def menu():
    gorevler = gorevleri_oku()
    while True:
        print("\n1. Görevleri Listele")
        print("2. Yeni Görev Ekle")
        print("3. Görev Düzenle")
        print("4. Görev Sil")
        print("5. Çıkış")
        secim = input("Seçiminiz: ").strip()

        if secim == "1":
            gorevleri_listele(gorevler)
        elif secim == "2":
            gorev_ekle(gorevler)
        elif secim == "3":
            gorev_duzenle(gorevler)
        elif secim == "4":
            gorev_sil(gorevler)
        elif secim == "5":
            print("Çıkış yapılıyor...")
            break
        else:
            print("Geçersiz seçim.")

menu()
```
